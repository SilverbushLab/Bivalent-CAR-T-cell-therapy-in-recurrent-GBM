```r
source('helpers.R')
```

```r
prepost <- readRDS("../../../ProcessedData/seurat/annotated_prepost_csf.rds")
```

```r
df <- subset(prepost@meta.data, cell_type %!in% c('CD4+ CAR T', 'CD8+ CAR T') & Day %in% c('D0','D7','D21'))
```

```r
result_df <- df %>%
 group_by(Day, Patient, cell_type) %>%
 summarise(n = n(), .groups = 'drop_last') %>%
 tidyr::complete(cell_type = unique(df$cell_type), fill = list(n = 0)) %>%
 mutate(Proportion = 100 * n / sum(n)) %>%
 dplyr::select(-n)

result_df_07 <- subset(result_df, Day %in% c('D0','D7'))
```

```r
colors <- cell_type_cols
```

```r
options(repr.plot.width = 2, repr.plot.height = 3, repr.plot.res=200)
p <- ggplot(data = subset(result_df_07, cell_type %in% c('Monocyte')), aes(y = Proportion, x = Day, fill = Day, color = cell_type)) + 
    geom_line(aes(group = Patient), col = "black",linewidth=0.1) +
    geom_point(aes(shape = Day), size = 2, position = position_dodge(width = 0.2), show.legend = FALSE) +
    facet_wrap(~ cell_type, nrow = 1, strip.position = "top") +
    labs(x = "", y = "Percentage") +
    scale_color_manual(values=colors) + ylim(0,30) +
    scale_shape_manual(values = c(20, 20)) + 
    theme_pubr() +  geom_signif(color='black', size=0.3, textsize=6,
        annotation = "*", 
        y_position = 27, xmin = 1, xmax = 2,
        tip_length = c(0.3, 0.1), 
      ) +
    theme(strip.background = element_blank(), strip.text.x = element_blank())
p
```

```r
pdf(file.path(panel.path, "4B_Monocyte.pdf"), width=2, height=3, family = "sans")
p
while (!is.null(dev.list()))  dev.off()
```
