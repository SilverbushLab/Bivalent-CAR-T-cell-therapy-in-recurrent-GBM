```r
prepost <- readRDS("../../../ProcessedData/seurat/annotated_prepost_csf.rds")
```

```r
cts = c('CD4+ T', 'CD4+ CAR T', 'CD8+ T', 'CD8+ CAR T', 'Cycling CD8+ T')
target <- subset(prepost, cell_type %in% cts) 
```

```r
target$cluster = 'CD8'
target$cluster[target$cell_type %in% c('CD4+ T', 'CD4+ CAR T')] = 'CD4'

target$group_by_cluster = 'CART' 
target$group_by_cluster[target$cell_type %!in% c('CD4+ CAR T', 'CD8+ CAR T')] = 'Endogenous'

target$group_by_transcript = 'CART' 
target$group_by_transcript[target$CAR_Transcript == 'CAR -'] = 'Endogenous'

target$sample = paste0(target$sample_id, '_', target$group_by_cluster)
```

```r
cd8.target = subset(target, cluster == 'CD8')
```

```r
genes = c('CD38','HLA-DRA','GNLY','IFI6', 'PRF1', 'CTLA4','LAG3','HAVCR2','PDCD1', 'IRF1', 'STAT1','KLRC1')
options(repr.plot.width=12, repr.plot.height=4, repr.plot.res=200)
p <- FeatureStatPlot(cd8.target, group.by = "group_by_cluster", ncol=6,
        stat.by = genes,   add_box=F, comparisons=list(c(1,2)), sig_label = "p.format",
        plot_type='violin', legend.position='none', add_point=F, pt.size=0.1, pt.alpha=1)  & 
    scale_fill_manual(values = c("CART" = "#D46F84", "Endogenous" = "#B294C7"))&
    xlab('') & ylab('') & theme(axis.text.x = element_blank()) & NoLegend()
p
```
